```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('books.csv', error_bad_lines=False)

pd.set_option('display.max_rows', 500)
```

```python
data = data.drop([3350, 4704, 5879, 8981], axis = 0)
```

```python
# Obras en inglés

data = data[data.language_code == 'eng']
data = data[data['language_code'].str.contains('eng')]
print(data.shape)
```

```python
# Top 10 books by average_rating

data.sort_values(by = 'average_rating', ascending = False)
```

```python
# Top 10 books

top_10 = data[0:10]

plt.figure(figsize = (18,8))
plt.barh(top_10["title"],top_10["average_rating"])
plt.title("Los 10 libros mejor valorados en Goodreads",fontdict = {"fontsize":20})
plt.savefig("top_10_books.jpg",dpi = 300)
plt.show()
```
